```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
df = pd.read_csv('world.csv', skiprows=4)
df.head((10))
```

```python
year = '2018'
```

```python
population_data = df[['Country Name', year]].dropna()
```

```python
population_data.sort_values(by=year, ascending=False, inplace=True)
```

```python
N = 30
top_countries = population_data.head(N)
```

```python
plt.figure(figsize=(12, 8))
plt.bar(top_countries['Country Name'], top_countries[year], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Population')
plt.title(f'Top {N} Countries by Population in {year}')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


well_known_countries = ['India', 'China', 'United Kingdom', 'United States', 'Germany', 'France', 'Mexico']

colors = ['blue', 'green', 'orange', 'red', 'purple', 'pink', 'cyan']

well_known_population_data = population_data[population_data['Country Name'].isin(well_known_countries)]


plt.figure(figsize=(14, 8))  # Increase the figure size for better readability
plt.bar(well_known_population_data['Country Name'], well_known_population_data[year], color=colors)
plt.xlabel('Country', fontsize=12)  
plt.ylabel('Population', fontsize=12)  
plt.title(f'Population of Well-Known Countries in {year}', fontsize=14)  # Adjust the font size of title
plt.xticks(rotation=45, ha='right', fontsize=10)  # Rotate x-axis labels and adjust font size
plt.tight_layout()
plt.show()
```
